```python
artist2tvec = {}
threshold_prob = 0.7
num_topics=10

def vectorize_topic_list(input_topic_list):
    t_vec = np.zeros((num_topics))
    t_vec[np.array(input_topic_list).T.astype(int)[0]] = np.array(input_topic_list).T[1]
    return np.where(t_vec>threshold_prob, t_vec, 0)

def turn_input_str_into_vec(input_str):
    input_bow = id2word.doc2bow(pf3.narrow_to_vocabulary(pf3.super_tokenize(input_str)))
    input_topics = pf3.lda_model.get_document_topics(input_bow, minimum_probability=threshold_prob)
    print(input_topics)
    return vectorize_topic_list(input_topics)

def create_topic_matrix():
    T_ = np.zeros((len(pf3.artists_list),num_topics))

    for i,rev_tok in enumerate(pf3.tokenized_reviews):
        rev_bow = id2word.doc2bow(rev_tok)
        artist2tvec[pf3.artists_list[i]] = rev_bow
        T_[i] = vectorize_topic_list(rev_bow)
    return T_

def get_artist_probs(t_mat, input_vec):
    return np.dot(T, test_input_vec.T).T
```

## create topic matrix that is compatible with searchcontroller tfidf

```python
a_to_i = {a:i for i,a in enumerate(cleaned_artist_names)}
```

```python
def fitted_matrix(input_topic_matrix, artist_to_idx):
    f_mat = np.zeros((len(artist_to_idx),input_topic_matrix.shape[1]))
    for i,a_row in enumerate(input_topic_matrix):
        a_name = pf3.artists_list[i]
        if a_name in artist_to_idx:
            f_mat[artist_to_idx[a_name]] = input_topic_matrix[i]
    return f_mat
```

```python
final_mat = fitted_matrix(topic_matrix, a_to_i)
```
